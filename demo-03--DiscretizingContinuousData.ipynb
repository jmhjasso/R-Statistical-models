### Discretization to approximate a continuous problem

__Quantization__:- Quantization is the process of constraining an input from a continuous or otherwise large set of values (such as the real numbers) to a discrete set (such as the integers). The terms quantization and discretization are often denotatively synonymous but not always connotatively interchangeable.


This function implements several basic unsupervised methods to convert a continuous variable into a categorical variable (factor) using different binning strategies. For convenience, a whole data.frame can be discretized (i.e., all numeric columns are discretized).

### Load library

```r
library(lambda.tools)
library(ggplot2)

options(repr.plot.width = 6, repr.plot.height = 5)
```

### Create continuous data 

```r
con_value <- seq(-20, 20, by=.5)

con_value
```

### Convert continuous data into discrete values by using `quantize` function
Note:- Here, number of discrete value are three (i.e   -1, 0, 1) by default

where, `quantize(x, bins=c(-1,0,1), metric=function(a,b) abs(a-b))`

```r
qnt_01 <- quantize(con_value)

qnt_01
```

### Converted in to data frame

```r
qnt_df_01 <- data.frame(con_value, qnt_01)

head(qnt_df_01, 5)
```

### Graphical representation of coverting countinuous data into descrete data points by using `ggplot`

```r
ggplot(qnt_df_01, aes(x = con_value, y = qnt_01)) +
    geom_point(size = 3, aes(color = factor(qnt_01)))
```

### Barplot of discrete values

```r
ggplot(
    data = qnt_df_01, 
    aes(factor(x = factor(qnt_01)))) + 
    geom_bar(col = 'black', fill = 'darkblue') 
```

```r
qnt_02 <- quantize(con_value, bins = c(-4, -3, -2, -1, 0,1, 2, 3, 4))

qnt_02
```

```r
qnt_df_02 <- data.frame(con_value, qnt_02)

head(qnt_df_02, 5)
```

```r
ggplot(qnt_df_02, aes(x = con_value, y = qnt_02)) +
 geom_point(size = 3, aes(colour = factor(qnt_02)))
```

### Create continuous data for sin function

```r
x <- seq(0, 2*pi, by=.1)

x
```

### Find the value of `sine` respective value of `x`

```r
y <- sin(x)

y
```

### Draw sine wave of given above values

```r
sin_wave <- qplot(x, y, geom="path", xlab="time", ylab="Sine wave") +
            geom_point(col = 'red', size = 3)+
            geom_hline(aes(yintercept = 0), size = 1, col = 'black')

sin_wave
```

### Create discrete value of sine into `0` and `1`

```r
qnt_03 <- quantize(sin(x), bins = c(-1, -.75, -.5, -.25, 0, .25, .5, .75, 1))

qnt_03
```

### Put it into dataframe

```r
qnt_df_03 <- data.frame(x, qnt_03)

head(qnt_df_03, 5)
```

### Graphical representation of discrete values of sine wave

Application of converting continuous values into descrete values is converting analog values to digital values 

https://en.wikipedia.org/wiki/Quantization_(signal_processing)

```r
ggplot(qnt_df_03, aes(x = x, y = qnt_03)) +
 geom_point(size = 3, aes(col = factor(qnt_03)))
```

```r
sin_wave + geom_step(aes(x = x, y = qnt_03), col = 'blue', size = .75)
```
