### Implementing of Derivative


Let's consider that a particles speed is given by this equation `S(t) = t^3 - 6 * t ^2 + 5 * t + 1`, and I want to find the rate of change of the body at particular time `t`.

Note :- S(t) in meter and t in sec.

http://www.oxnotes.com/movement-and-position.html

https://www.dummies.com/education/math/calculus/how-to-analyze-position-velocity-and-acceleration-with-differentiation/

### Equation for motion of the body `S(t) = t^3 - 6 * t ^2 + 5 * t + 1`

```r
S_t <- makeFun(t^3 - 6 * t^2 + 5 * t + 1 ~ t)

S_t
```

### Graphical view of partical is moving on equation.

```r
plotFun(S_t, xlim = range(0, 5), ylim = range(-20, 10), lwd = 3, col = 'blue',
        ylab = 'Distance', xlab = 'time', main = 'S(t) = t^3 - 6 * t^2 + 5 * t + 30')
```

### The rate of change of body is calculated by `V(t) =  d(S(t))/d( t)` i.e `Deriv(S(t))`.

```r
V_t <- Deriv(S_t)

V_t
```

### The velocity of particular at t = 5 sec

```r
V_t(5)
```

### Graphical view of the rate of change of distance or `S(t)`.

```r
plotFun(V_t, xlim = range(0, 4), ylim = range(-8, 4), lwd = 3, col = 'red',
        ylab = 'Velocity', xlab = 'time', main = 'V(t) = 3 * t^2 - 12 * t + 5')
```

### Calculate the rate of change of velocity i.e `acceleration`.
Note:- `acceleration = d(V(t))/ d(t)`  i.e `a = Deriv(V(t))`.

```r
V_t <- makeFun(5 + t * (3 * t - 12) ~ t)

V_t
```

```r
acc <- Deriv(V_t)

acc
```

### Acceleration at `t = 5 sec` 
Note:- the unit of acceleration will `m/s^2`

```r
acc(5)
```

### Graphical view of `acceleration` graph

```r
plotFun(acc, xlim = range(-2, 8), ylim = range(-18, 25), lwd = 3, col = 'red',
        ylab = 'Acceleration', xlab = 'time', main = '6 * t - 12')
```
