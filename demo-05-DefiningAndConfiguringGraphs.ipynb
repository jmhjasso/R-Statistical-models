### Undirected graph

```r
g1 <- graph(edge = c(1,2, 2,3, 2,4, 3,1, 4,1), n = 4, directed = F)

plot(g1, vertex.size = 40)
```

### Directed graph

```r
g2 <- graph(edge = c(1,2, 2,1, 2,3, 2,4, 3,1, 4,1), n = 4, directed = T)

plot(g2, vertex.size = 50)
```

### Connected Graph

```r
g3 <- graph(edge = c(1,2, 2,3, 3,7, 7,5, 5,1, 5,6, 7,4), n = 7, directed = T)

plot(g3, vertex.size = 30)
```

### Unconnected graph

```r
g4 <- graph(edge = c(1,4,  2,4,  3,4,  7,5,  5,1), n = 9, directed = T)

plot(g4, vertex.size = 30)
```

### Named graph

```r
g5 <- graph( c('IND','USA', 'IND','USA', 'USA','AUS',  
               'AUS','IND', 'IND','CHN', 'CHN','IND',
               'JPN','IND', 'USA','CHN', 'CHN','USA'), directed = T)

plot(g5, vertex.size = 50, vertex.label.cex = 1)
```

### Isolated graph

```r
g6 <- graph(c('Ford','Suzuki',  'Suzuki','Tata',  'Tata','Nissan',
              'Nissan','Toyota',  'Nissan','Nissan',  'Toyota','Ford'),
            
           isolates = c('Honda', 'Volvo'))

plot(g6, vertex.size = 50, edge.arrow.size = 1, vertex.color = 'green', vertex.frame.color = 'red',
     vertex.label.color = 'black', vertex.label.cex = .75, vertex.lebel.dist = 4, edge.curved = .2)
```

### Edges

```r
E(g6)
```

### Vertices

```r
V(g6)
```

### Matrix representation of graph

```r
g6[]
```

### Select which node is connected to a particular node

```r
V(g6)$country = c("USA", "Japan", "India", "Japan", "Japan", "Japan", "Sweden")

vertex_attr(g6)
```

```r
E(g6)$relationship = c("Research", "Alliance", "Partnership", "Research", "Autoparts", "Alliance")

edge_attr(g6)
```

```r
g6 <- set_graph_attr(g6, "about", "Car manufacturing companies")

g6 <- set_graph_attr(g6, "data", "Relationships")

graph_attr(g6)
```

### Another way to create small graph

#### 1. Undirected graph

```r
plot(graph_from_literal(Subaru--Suzuki, 
                        Suzuki---Tata, 
                        Tata--Tesla, 
                        Tesla------Subaru, 
                        Tesla--Suzuki ),
     vertex.size = 65, vertex.color = 'yellow')
```

#### 2. Directed graph

```r
plot(graph_from_literal(Subaru--+Suzuki, 
                        Suzuki+---Tata, 
                        Tata--+Tesla, 
                        Tesla+--Subaru, 
                        Tesla+--+Suzuki),
    vertex.size = 65, vertex.color = 'yellow')
```

### Specific graphs and graph models

### Create empty graph by using `make_empty_graph` function

```r
empty_g <- make_empty_graph(30)

plot(empty_g, vertex.size = 10, vertex.label = NA)
```

### Create full graph by using `make_full_graph` function

```r
full_g <- make_full_graph(20)
plot(full_g, vertex.size = 10, vertex.label = NA)
```

### Create star graph by using `make_star` function

```r
star_g <- make_star(30)
plot(star_g, vertex.size = 10, vertex.label = NA)
```

### Create tree graph by using `make_tree` function

#### 1. Undirected

```r
tree_g <- make_tree(30, children = 3, mode = 'undirected')

plot(tree_g , vertex.size = 10, vertex.label = NA)
```

#### 2. Directed graph

```r
tree_g <- make_tree(30, children = 3, mode = 'out')

plot(tree_g , vertex.size = 10, vertex.label = NA)
```

#### 3. Directed graph

```r
tree_g <- make_tree(30, children = 3, mode = 'in')

plot(tree_g , vertex.size = 10, vertex.label = NA)
```

### Create ring graph by using `make_ring` function

```r
ring_g <- make_ring(30)
plot(ring_g, vertex.size = 10, vertex.label = NA)
```

### Erdos-Renyi random graph model
(‘n’ is number of nodes, ‘m’ is the number of edges).

```r
erdos_renyi_g <- sample_gnm(n = 50, m = 35)
plot(erdos_renyi_g, vertex.size = 10, vertex.label = NA)
```

### Create sample world graph by using `sample_smallworld` function

#### 1. this is in 2D

```r
small_world_g <- sample_smallworld(dim = 2, size = 10, nei = 1, p = 0.1)

plot(small_world_g, vertex.size = 10, vertex.label = NA, layout = layout_in_circle)
```

```r
small_world_g <- sample_smallworld(dim = 2, size = 10, nei = 3, p = 0.1)

plot(small_world_g, vertex.size = 10, vertex.label = NA, layout = layout_in_circle)
```

#### 2. This is in 3D

```r
small_world_g <- sample_smallworld(dim = 3, size = 10, nei = 1, p = 0.1)

plot(small_world_g, vertex.size = 10, vertex.label = NA, layout = layout_in_circle)
```
