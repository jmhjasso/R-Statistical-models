## Integration

### Create  a function which gives `constant` value only by using `manual` function

```r
f0 <- function(x){
    m = 0; b = 3;
    m * x + b;
}

f0
```

### graphical view of constant value

```r
plot(f0, xlim = c(0, 5), ylim = c(0, 6), col = 'blue', lwd = 3, main = 'f(x) = m * x + b')
```

### Integration of constant value by using `integrate` function and make it `vectorize`

```r
I0 <- function(x) {integrate(f0, lower=0, upper=x)$value}

vect_fun = Vectorize(I0, vectorize.args='x')
```

### Graphical view of above integrated function

```r
t <- seq(0, 5, .5)

plot(t, vect_fun(t), xlim = c(0, 5), ylim = c(0, 20), type = 'l', lwd = 3,
     col = 'green', ylab = 'f(x)', xlab = 'x', main = 'f(x) = m / 2 * x^2 + b*x + c')
```

### Implementation of integration

Note:- Here, we take equation of acceleration of above problem. Now we are going to integrate to get velocity `V(t)` i.e `antiD(V(t))` 

### Create Acceleration function

```r
Acc <- function(x){6 * x - 12}
Acc
```

### Graphical view of Acceleration function

```r
plot(Acc, xlim = c(0, 8), ylim = c(-15, 50), col = 'blue', lwd = 3,
     ylab = 'Acceleration', xlab = 'time', main = '6*x - 12')
```

### Integrated of `acceleration` function so you can get `velocity`

```r
V_t <- function(x) {integrate(Acc, lower=0, upper=x)$value}

V_fun = Vectorize(V_t, vectorize.args='x')
```

### graphical view of velocity

#### These are values of spped when time falls between `0 to 5`

```r
t <- seq(0, 5, 0.1)

plot(t, V_fun(t), xlim = c(0, 5), ylim = c(-12, 20), type = 'l', col = 'green', lwd = 3, 
     main = '3*x^2 + 12*x + C')
```

### Now, again create `velocity function` manualy

```r
V_t <- function(x){3 * x**2 - 12 * x}

V_t
```

### Graphical view of `velocity` 

```r
plot(V_t, xlim = c(0, 5), ylim = c(-15, 20), col = 'blue', lwd = 3,
     ylab = 'Acceleration', xlab = 'time', main = '3x^2 -12x')
```

### Integrate `velocity` function to get speed

```r
S_t <- function(x) {integrate(V_t, lower=0, upper=x)$value}

S_fun = Vectorize(S_t, vectorize.args='x')
```

```r
t <- seq(0, 5, 0.1)

plot(t, S_fun(t), xlim = c(0, 5), ylim = c(-50, 2), type = 'l', col = 'green', lwd = 3)
```
